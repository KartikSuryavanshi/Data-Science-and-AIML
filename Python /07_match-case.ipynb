```python
code=2000  # for improving redeability we use match-case
```

```python
match code:
    case 200:
        print("response ok ")
        print("good job")
    case 400:
        print("bad response")
    case 404:
        print("not found")
    case _:#default case
        print("nothing matched")
print("end ")
```

```python
point = (3, 4)

match point:
    case (0, 0):
        print("Origin")
    case (x, 0):
        print(f"X-Axis at {x}")
    case (0, y):
        print(f"Y-Axis at {y}")
    case (x, y):
        print(f"Point at ({x}, {y})")
```
