```python
l =['apples','banana','tomato']
for item in l:
    if(item=='banana'):
        break#stop this loop right here
    print(item)
```

```python
l =['apples','banana','tomato']
for item in l:
    if(item=='banana'):
        continue#skip this particular iteration of the loop that why banana is skipped here
    print(item)
```
