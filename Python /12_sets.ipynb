```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
```

```python
set1.union(set2)
```

```python
set1.intersection(set2)
```

```python
set1
```

```python
set2
```

```python
sub={1,3}
```

```python
sub.issubset(set1)
```

```python
set1.issuperset(sub)#ie all elements of sub are there in set1 -- superset
```

```python
# Union – combines all unique elements
print(set1 | set2)            # {1, 2, 3, 4, 5, 6}
print(set1.union(set2))       # same result
 
# Intersection – common elements
print(set1 & set2)            # {3, 4}
print(set1.intersection(set2))# same result
```

```python
# Difference – in set1 but not in set2
print(set2 - set1) #Elements in set2 but not in set1.ie it only takes element of set2
print(set1 - set2) #Meaning: Elements in set1 but not in set2.ie it only takes element of set1
print(set1.difference(set2))  # same result
 
# Symmetric Difference – in either set, but not both
print(set1 ^ set2)                     # {1, 2, 5, 6}
print(set1.symmetric_difference(set2))# same result
```
