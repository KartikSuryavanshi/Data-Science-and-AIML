```python
squares = [x**2 for x in range(5)]
print(squares)
```

```python
evens = [x for x in range(10) if x % 2 == 0]
print(evens) 
```

```python
numbers = ["Even" if x % 2 == 0 else "Odd" for x in range(5)]
print(numbers)
```

```python
pairs = [(x, y) for x in range(2) for y in range(3)]
print(pairs)
```

```python
words = ["hello", "world", "python"]
upper_words = [word.upper() for word in words]
print(upper_words)
```

```python
matrix = [[1, 2], [3, 4], [5, 6]]
flattened = [num for row in matrix for num in row]
print(flattened)
```

```python
unique_numbers = {x for x in [1, 2, 2, 3, 4, 4]} #set comprehension
print(unique_numbers)
```

```python
squared_dict = {x: x**2 for x in range(5)} #dictionary comprehension
print(squared_dict) 
```

```python
import time #List comprehensions are generally faster than loops because they are optimized internally by Python.

# Using a for loop
start = time.time()
squares_loop = []
for x in range(10**6):
    squares_loop.append(x**2)
print("Loop time:", time.time() - start)

# Using list comprehension
start = time.time()
squares_comp = [x**2 for x in range(10**6)]
print("List Comprehension time:", time.time() - start)
```
